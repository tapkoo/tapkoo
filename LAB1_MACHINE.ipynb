```python
import numpy as ny
import matplotlib.pyplot as ppt
from sklearn.cluster import KMeans
import statistics
```

```python
from sklearn.datasets.samples_generator import make_blobs
X, y_true = make_blobs(n_samples=100, centers=6,
                       cluster_std=0.60, random_state=6)
ppt.scatter(X[:, 0], X[:, 1], s=50);
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=6)
kmeans.fit(X)
y_kmeans = kmeans.predict(X)
ppt.scatter(X[y_kmeans == 0, 0], X[y_kmeans == 0, 1], s = 20, c = 'orange', label = 'Cluster1')
print('Orange cm: ', (statistics.mean(X[y_kmeans == 0, 1])), '| median:', (statistics.median(X[y_kmeans == 0, 1])))
ppt.scatter(X[y_kmeans == 1, 0], X[y_kmeans == 1, 1], s = 20, c = 'blue', label = 'Cluster2')
print('Blue cm: ', (statistics.mean(X[y_kmeans == 1, 1])), '| median:', (statistics.median(X[y_kmeans == 1, 1])))
ppt.scatter(X[y_kmeans == 2, 0], X[y_kmeans == 2, 1], s = 20, c = 'green', label = 'Cluster3')
print('Green cm: ', (statistics.mean(X[y_kmeans == 2, 1])), '| deviation:', (statistics.stdev(X[y_kmeans == 2, 1])))
ppt.scatter(X[y_kmeans == 3, 0], X[y_kmeans == 3, 1], s = 20, c = 'brown', label = 'Cluster4')
print('Brown cm: ', (statistics.mean(X[y_kmeans == 3, 1])), '| deviation:', (statistics.stdev(X[y_kmeans == 3, 1])))
ppt.scatter(X[y_kmeans == 4, 0], X[y_kmeans == 4, 1], s = 20, c = 'red', label = 'Cluster5')
print('Red cm: ', (statistics.mean(X[y_kmeans == 4, 1])), '| median:', (statistics.median(X[y_kmeans == 4, 1])), '| deviation:', (statistics.stdev(X[y_kmeans == 4, 1])))
ppt.scatter(X[y_kmeans == 5, 0], X[y_kmeans == 5, 1], s = 20, c = 'cyan', label = 'Cluster6')
print('Cyan cm: ', (statistics.mean(X[y_kmeans == 5, 1])), '| median:', (statistics.median(X[y_kmeans == 5, 1])), '| deviation:', (statistics.stdev(X[y_kmeans == 5, 1])))
ppt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s = 80, c = 'black', label = 'Centroids')
ppt.show()
```
